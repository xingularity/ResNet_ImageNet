```python
# Import what we need to read image and resize
import cv2
import os
```

```python
# relative path to the classes of images
base_dir = './imagenet_images'
classes = os.listdir('./imagenet_images')
```

```python
# real all file path into a 2D Python list
files = [[os.path.join(x, y) for y in os.listdir(x)] for x in [os.path.join(base_dir, cls) for cls in classes]]
```

```python
# resize the image and write into files
def resize_img(path, file_name):
    img = cv2.imread(path, cv2.IMREAD_COLOR)
    cv2.imwrite(file_name, cv2.resize(img, (224,224), interpolation=cv2.INTER_AREA))
```

```python
# do resizing images and put into folder '224X224'
[[resize_img(y, y.replace('imagenet_images', '224X224')) for y in x] for x in files]
```

```python
import numpy as np
import h5py
from sklearn.model_selection import train_test_split
```

```python
def read_image(path):
    return cv2.imread(path, cv2.IMREAD_COLOR)
```

```python
image_base_dir = './224X224'
classes = os.listdir(image_base_dir)
classes = [x for x in classes if not x.startswith('.')]
for cls in classes:
    print(cls)
    dir_path = os.path.join(image_base_dir, cls)
    files = os.listdir(dir_path)
    images = [read_image(os.path.join(dir_path, x)) for x in files]
    images_one_cls = np.concatenate([x.reshape(1,224, 224, 3) for x in images if x is not None])
    y = np.ndarray(images_one_cls.shape[0])
    
    X_train, X_test, Y_train, Y_test = train_test_split(images_one_cls, y, test_size=0.2)
    print(X_train.shape,  X_test.shape, Y_train.shape, Y_test.shape)
    with h5py.File('x_train.h5', 'a') as h5f:
        h5f.create_dataset(cls, data=X_train)
    with h5py.File('x_test.h5', 'a') as h5f:
        h5f.create_dataset(cls, data=X_test)
```
